```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
age_cost = df[['age', 'charges']]
```

```python
plt.scatter(age_cost['age'], age_cost['charges'])
plt.show()
model = LinearRegression()
```

```python
model.fit(age_cost[['age']], age_cost['charges'])
```

```python
predictions = model.predict(age_cost[['age']])
```

```python
plt.scatter(age_cost['age'], age_cost['charges'])
plt.plot(age_cost['age'], predictions, color='red')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
